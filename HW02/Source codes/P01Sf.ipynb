```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# load the dataset
df = pd.read_csv('HW02P01Sf.csv')
df.head()
```

```python
# convert categorical features
label_encoder = LabelEncoder()
df['F1 '] = label_encoder.fit_transform(df['F1 '])
df['F2 '] = label_encoder.fit_transform(df['F2 '])
df['F3 '] = label_encoder.fit_transform(df['F3 '])
df['F4 '] = label_encoder.fit_transform(df['F4 '])
df['F5 '] = label_encoder.fit_transform(df['F5 '])
df.head()
```

```python
df = df.drop(columns=['Id '])
```

```python
df.head()
```

```python
X = df.drop(['Target'], axis=1)
y = df['Target']
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier()
tree_model.fit(X, y)
```

```python
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(tree_model, feature_names=X.columns, class_names=['Mines', 'Plus'], filled=True, rounded=True, special_characters=True)
graphviz.Source(dot_data)
```
