```python
def check(df):
    l=[]
    columns=df.columns
    for col in columns:
        dtypes=df[col].dtypes
        nunique=df[col].nunique()
        sum_null=df[col].isnull().sum()
        l.append([col,dtypes,nunique,sum_null])
    df_check=pd.DataFrame(l)
    df_check.columns=['column','dtypes','nunique','sum_null']
    return df_check
```

```python
check(lp)
```

```python
#d
from sklearn.preprocessing import LabelEncoder
cols=['Company','Product','TypeName','ScreenResolution','Cpu','Memory','Gpu','OpSys']
la= LabelEncoder()
for i in cols:
    lpc[i] = la.fit_transform(lpc[i])
lpc
```

```python
lpf = lpc.copy()
y=lpf.pop('Price_euros')
X=lpf.drop(['Product'], axis=1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
y = scaler.fit_transform(y.to_frame())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=1)
X_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score


k = 10
kf = KFold(n_splits=k)

degrees = range(1, 8)
mse_mean = np.zeros(len(degrees))
mse_std = np.zeros(len(degrees))

for i, degree in enumerate(degrees):
    mse = []
    for train_idx, val_idx in kf.split(X):
        X_trainp, y_trainp = X[train_idx], y[train_idx]
        X_val, y_val = X[val_idx], y[val_idx]
        
        model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
        
        model.fit(X_trainp, y_trainp)
        
        y_pred = model.predict(X_val)
        mse.append(mean_squared_error(y_val, y_pred))
    print(mse)
    mse_mean[i] = np.mean(mse)
    mse_std[i] = np.std(mse)
    print(" D = {} Complete training.".format(degree))

plt.errorbar(degrees, mse_mean, yerr=mse_std, fmt='o-', color='b', ecolor='r')
plt.title('MSE vs Polynomial Degree')
plt.xlabel('Polynomial Degree')
plt.ylabel('MSE')
plt.show()
```
